```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tr_estmtr import tr_estmtr
```

```python
dim = 100
B = np.random.uniform(0,1,size=(dim, dim))

print(B)

A = B@B.T

print(A)
print(f"trace : {np.trace(A)}")
```

```python
sum = 0
data = []
for i in range(10):
    x = tr_estmtr(A)
    data = np.append(data,x)

plt.hist(data, bins=30, color='blue', alpha=0.7)

plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```
